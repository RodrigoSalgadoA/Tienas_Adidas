```python
# proyecto_adidas_sucursales_imprimir.py
# Requisitos: pandas, numpy, matplotlib, seaborn, scikit-learn
# Instalar: pip install pandas numpy matplotlib seaborn scikit-learn

import logging
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 120
```

```python
# 1) Configuración
INPUT_CSV = "Davivienda_Dataset__Sucursales_A_adidas_.csv"
RANDOM_STATE = 42

logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s: %(message)s")
logger = logging.getLogger(__name__)
```

```python
# 2) Cargar datos
df = pd.read_csv(INPUT_CSV, low_memory=False)
print("Filas, columnas:", df.shape)
print("Columnas:", df.columns.tolist())
```

```python
# 3) Limpieza y conversiones
def clean_and_cast(df):
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]
    # Campos numéricos conocidos
    numeric_cols = [
        "Tiempo_Pago_Horas", "Ingresos_Mensuales", "Frecuencia_Transacciones",
        "Costo_Transporte_Pago", "Numero_Cuentas_Bancarias", "Sucursales_10KM",
        "Satisfaccion_Servicios_Financieros", "Edad"
    ]
    for c in numeric_cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c].astype(str).str.replace(r'[^\d\.\-]', '', regex=True), errors='coerce')
    # Flags binarios a 0/1
    for c in ['Uso_Tecnologia_Movil', 'Tenencia_Telefono_Celular', 'Acceso_Servicios_Financieros']:
        if c in df.columns:
            df[c] = df[c].astype(str).str.lower().map(lambda x: 1 if x in ['si','sí','yes','y','1','true','t'] else 0)
            df[c] = pd.to_numeric(df[c], errors='coerce').fillna(0).astype(int)
    # Normalizar Preferencia_Pago text
    if 'Preferencia_Pago' in df.columns:
        df['Preferencia_Pago'] = df['Preferencia_Pago'].astype(str).str.strip().replace('nan', np.nan)
    return df

df = clean_and_cast(df)
print("\nNulos por columna:")
print(df.isna().sum())
```

```python
# 4) Seleccionar unidad de agrupación (Region_Geografica o fallback)
group_col = 'Region_Geografica' if 'Region_Geografica' in df.columns else None
if group_col is None:
    df['Region_Auto'] = pd.qcut(df.get('Ingresos_Mensuales', pd.Series([0]*len(df))), q=5, duplicates='drop').astype(str)
    group_col = 'Region_Auto'
    logger.warning("No existe Region_Geografica; se creó Region_Auto por quintiles de ingresos")

print(f"\nUsando '{group_col}' como unidad de agregación.")

```

```python
# 5) Cálculo de KPIs por región
def compute_kpis_region(df, group_col):
    agg = {}
    if 'Ingresos_Mensuales' in df.columns:
        agg['IngresosTotales'] = ('Ingresos_Mensuales', 'sum')
        agg['IngresoPromedio'] = ('Ingresos_Mensuales', 'mean')
    if 'Frecuencia_Transacciones' in df.columns:
        agg['FrecuenciaPromedio'] = ('Frecuencia_Transacciones', 'mean')
        agg['FrecuenciaTotal'] = ('Frecuencia_Transacciones', 'sum')
    if 'Tiempo_Pago_Horas' in df.columns:
        agg['TiempoPagoPromedio'] = ('Tiempo_Pago_Horas', 'mean')
    if 'Edad' in df.columns:
        agg['EdadPromedio'] = ('Edad', 'mean')
    if 'Sucursales_10KM' in df.columns:
        agg['Sucursales_10KM_Mean'] = ('Sucursales_10KM', 'mean')
    if 'Numero_Cuentas_Bancarias' in df.columns:
        agg['CuentasPromedio'] = ('Numero_Cuentas_Bancarias', 'mean')
    if 'Tenencia_Telefono_Celular' in df.columns:
        agg['PenetracionMovil'] = ('Tenencia_Telefono_Celular', 'mean')
    if 'Uso_Tecnologia_Movil' in df.columns:
        agg['UsoMovilProm'] = ('Uso_Tecnologia_Movil', 'mean')
    if 'Satisfaccion_Servicios_Financieros' in df.columns:
        agg['SatisfaccionPromedio'] = ('Satisfaccion_Servicios_Financieros', 'mean')

    kpis = df.groupby(group_col).agg(**agg).reset_index()
    if 'IngresosTotales' in kpis.columns and 'FrecuenciaTotal' in kpis.columns:
        kpis['Ingreso_por_Transaccion'] = kpis['IngresosTotales'] / kpis['FrecuenciaTotal'].replace(0, np.nan)
    return kpis

kpis = compute_kpis_region(df, group_col)
```

```python
# 6) Imprimir dataframes resumidos
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)
print("\n--- Muestra del dataframe original (head) ---")
print(df.head(10))

print("\n--- Descripción estadística (original) ---")
print(df.describe(include='all').transpose().head(20))

print("\n--- KPIs por región (head) ---")
print(kpis.head(20))

print("\n--- Estadísticas resumen de KPIs ---")
print(kpis.describe().transpose())
```

```python
# 7) Preparar y ejecutar clustering sobre features disponibles
candidate_features = [
    'IngresosTotales', 'IngresoPromedio', 'FrecuenciaPromedio',
    'TiempoPagoPromedio', 'EdadPromedio', 'Sucursales_10KM_Mean',
    'CuentasPromedio', 'PenetracionMovil', 'UsoMovilProm', 'SatisfaccionPromedio'
]
features_for_clust = [f for f in candidate_features if f in kpis.columns]

if len(features_for_clust) < 1:
    print("\nNo hay features numéricas suficientes para clustering. Se omite clustering.")
    kpis['Cluster'] = 0
else:
    X = kpis[features_for_clust].replace([np.inf, -np.inf], np.nan).fillna(0)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(X)
    max_k = min(6, max(2, len(kpis)-1))
    best_k = 2
    best_score = -1
    for k in range(2, max_k+1):
        km = KMeans(n_clusters=k, random_state=RANDOM_STATE, n_init=10)
        labels = km.fit_predict(Xs)
        if len(np.unique(labels)) < 2:
            continue
        score = silhouette_score(Xs, labels)
        if score > best_score:
            best_score = score
            best_k = k
    km_final = KMeans(n_clusters=best_k, random_state=RANDOM_STATE, n_init=20)
    kpis['Cluster'] = km_final.fit_predict(Xs)
    print(f"\nClustering aplicado: k={best_k}, silhouette={best_score:.4f}")
    print("\nConteo por cluster:")
    print(kpis['Cluster'].value_counts())
```

```python
# 8) Imprimir resumen de clustering y primeras filas
print("\n--- KPIs con Cluster (head) ---")
print(kpis.head(20))

cluster_summary = kpis.groupby('Cluster').agg(
    Regiones=('Region_Geografica' if 'Region_Geografica' in kpis.columns else group_col, 'count'),
    IngresosTotalesCluster=('IngresosTotales', 'sum'),
    IngresoPromedioCluster=('IngresoPromedio', 'mean')
).reset_index()
print("\n--- Resumen por cluster ---")
print(cluster_summary)
```

```python
# 9) Visualizaciones en pantalla (plt.show())
def quick_plots(kpis, df):
    # Top regiones por IngresosTotales
    if 'IngresosTotales' in kpis.columns:
        top = kpis.sort_values('IngresosTotales', ascending=False).head(10)
        plt.figure(figsize=(9,6))
        sns.barplot(data=top, x='IngresosTotales', y=top.columns[0] if 'Region_Geografica' in top.columns else top.columns[0], palette="viridis")
        plt.title("Top regiones por ingresos totales")
        plt.tight_layout()
        plt.show()

    # Conteo por cluster
    if 'Cluster' in kpis.columns:
        plt.figure(figsize=(6,4))
        sns.countplot(data=kpis, x='Cluster', palette="Set2")
        plt.title("Conteo de regiones por cluster")
        plt.tight_layout()
        plt.show()

    # Histograma de Ingresos mensuales (nivel individuo)
    if 'Ingresos_Mensuales' in df.columns:
        plt.figure(figsize=(8,4))
        sns.histplot(df['Ingresos_Mensuales'].dropna(), bins=40, kde=True)
        plt.title("Distribución de Ingresos Mensuales (individuos)")
        plt.tight_layout()
        plt.show()

quick_plots(kpis, df)

print("\nProceso completado. Todos los resultados se imprimieron y las gráficas se mostraron en pantalla.")
```
